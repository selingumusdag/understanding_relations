```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
```

```python
weather=pd.read_csv('./data/weatherHistory.csv')
weather.head()
```

```python
df = weather[['Humidity','Wind Speed (km/h)','Temperature (C)', 'Apparent Temperature (C)']]
#target value
Y = df['Temperature (C)']- df['Apparent Temperature (C)']
#explanatory variables
X = df[['Humidity','Wind Speed (km/h)']]
```

```python
X=sm.add_constant(X)
results=sm.OLS(Y,X).fit()
results.summary()
```

```python
df['hum_wind']=df['Humidity']*df['Wind Speed (km/h)']
```

```python
X = df[['Humidity','Wind Speed (km/h)','hum_wind']]
X=sm.add_constant(X)

results=sm.OLS(Y,X).fit()
results.summary()
```
